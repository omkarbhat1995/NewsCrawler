```python
from textblob import TextBlob
import requests
from bs4 import BeautifulSoup
list_of_keywords = combine#['medical+vulnerable+hack','hospital+vulnerable+hack','nhs+hack','hospital+hack','medical+hack','antivirus','computer+virus','malware','securing','secure','passive+attack','active+attacks','integrity','security','safety','vulnerability','Phishing','Ransomware','backdoor','Denial+of+Service','Dos','Distributed Denial of Service','DDos','hack','Hacks','spyware','worm','trojan','adware','spam','privacy','bug','crimeware','confidentiality','hijack','hoax','infection','Identity Theft','man-in-the-middle','MITM','trackware','attacker','nonsecure','authentication','penetration','cryptography','crypto','private+key','threat','targeted+attack','encryption','decipher','decoy+server','cipher','cryptanalysis','intrusion','malicious','data breach','adversarial','perturbation','misuse','rootkit','exploit','password','firewall','root+access','unauthorized','buffer+overflow','sniffer','snooping','software+security'] #search keywords
links=[]
class Analysis:
    def __init__(self,term):
        self.term=term
        self.subjectivity=0
        self.url='https://www.google.com/search?q={0}&source=lnms&tbm=nws'.format(self.term)
        
    def run(self):
        response=requests.get(self.url)
        soup=BeautifulSoup(response.text,'html.parser')
        headline_results=soup.find_all('a')
        for h in headline_results:
            links.append(h)
            blob=TextBlob(h.get_text())
            self.subjectivity+=blob.sentiment.subjectivity/ len(headline_results)
qw=[]
for t in list_of_keywords:
    a=Analysis(t)
    a.run()
    qw.append(a.subjectivity)
```

```python
list1=[]
for j in links:
    if "q=https://" in str(j):
        list1.append(j)
rty=[]
for i in range(len(list1)):
    if i%2==0:
        st=str(list1[i])
        st=st[10:]
        t=st.find('=')
        t1=st.find('&amp')
        at=st[t+1:t1]
        rty.append(at)
#now use "at" to help us find helpful data
```

```python
rty2=[ [ None for y in range( len(comp_keywords)+3 ) ] for x in range( len(rty) ) ]
for j in range(int(len(rty)/5)):
    rty2[j][0]=rty[j]
for j in range(len(rty)):
    response=requests.get(rty2[j][0])        
    soup=BeautifulSoup(response.text,'html.parser')
    paras=soup.find_all('p')
    text=""
    for t in paras:
        webtext=t.get_text()
        text=text+webtext
        text=text.replace("\n","")
        text=text.replace("\t","")
        text=text.replace("\''","''")
    rty2[j][1]=text
    if j%50==0:
        print(j)
```
